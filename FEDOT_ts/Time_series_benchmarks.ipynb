```python
from matplotlib import pyplot as plt
from pylab import rcParams
rcParams['figure.figsize'] = 11, 4
```

```python
from autots.datasets import load_monthly, load_hourly, load_daily, load_weekly, load_yearly
long = True

df_monthly = load_monthly(long=long)
df_hourly = load_hourly(long=long)
df_daily = load_daily(long=long)
df_weekly = load_weekly(long=long)
df_yearly = load_yearly(long=long)
```

```python
# Define function
def print_ts(df):
    try:        
        variables = list(df.series_id.unique())
        try:
            names = list(df.series_name.unique())
        except Exception:
            names = [' ']*len(variables)

        for v, name in zip(variables, names):
            df_var = df[df['series_id'] == v]

            plt.plot(df_var['datetime'], df_var['value'])
            plt.ylabel(str(name))
            plt.xlabel('Date')
            plt.title(v)
            plt.grid()
            plt.show()
    except Exception:
        pass
```

```python
dataframes = [df_monthly, df_hourly, df_daily, df_weekly, df_yearly]
dataframe_names = ['monthly', 'hourly', 'daily', 'weekly', 'yearly']

for df, df_name in zip(dataframes, dataframe_names):
    print(f'\n{df_name}\n')
    print_ts(df)
```
